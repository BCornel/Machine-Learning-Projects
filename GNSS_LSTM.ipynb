# Description of data

This dataset is a GNSS time series of position over time. The time column is measured in decimal-year and represents one day (unless there is a break in measurement). The GNSS receiver measures in easting, northin and vertical. For this project I will look at the vertical, which shows the subsidence, or the sinking of the land. 

The purpose is to get my foot in the door with using different methods to capture trend information and make predictions in the motion of the local crustal deformation. Currently, this will look at using the LSTM model with a univariate case (one variable i.e. change in position) and multivariate (2 variables i.e. change in position and time). 

The LSTM looks back into time a set number of days, in this program these are the hops, and then predicts those days into the future. I loop through a series of hops to understand the effect this has on the learned model. Currently, it goes through 7, 14, 21 and 28 days.

# Univarate LSTM case (just space with no time dependancy)

```python
import pandas as pd
import matplotlib.pyplot as plt

#Load the csv data into pandas dataframe
station='MRHK'
# Read in data only taking the vertical measurements
df = pd.read_csv('./MRHK_GOM20_neu_cm.col',
                 header=0, 
                 delim_whitespace=True,
                 usecols=[0,3])

print(df.head())
# Visualize the dataset
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df['Decimal-Year'],
        df['UD(cm)'],
        'k.')
ax.set_xlabel('Time (day)')
ax.set_ylabel('Z (cm)')
ax.set_title('GNSS Station: %s'%station)
```

```python
# training/test split 
import numpy as np
test_size=int(df.shape[0] * 0.3)
X_train=np.array(df['UD(cm)'].iloc[test_size:]).reshape(-1,1)
X_test=np.array(df['UD(cm)'].iloc[:test_size]).reshape(-1,1)


from sklearn.preprocessing import MinMaxScaler
# Convert from 0 to 1 (normalize)
scaler = MinMaxScaler()
scaler.fit(X_train)
scaler_train = scaler.transform(X_train)
# Remember to scale the test data 
scaler_test = scaler.transform(X_test)
scaler_train.shape[0]

def univariate_lstm(df, n_days):
  hops=n_days   # this is in days
  total_len=scaler_train.shape[0]
  X_train=[]
  y_train=[]

  # Reshape the array for LSTM model
  for i in range(hops,total_len):
    X_train.append(scaler_train[i-hops:i])  # x_train is the past data i-hops
    y_train.append(scaler_train[i])         # y data is current data i
  X_train=np.array(X_train)
  y_train=np.array(y_train)
  print(X_train.shape)
  print(y_train.shape)


  # Do the same for test data
  n_features=1
  X_test=[]
  y_test=[]

  for i in range(hops, scaler_test.shape[0]):
    X_test.append(scaler_test[i-hops:i])
    y_test.append(scaler_test[i])
  X_test=np.array(X_test)
  y_test=np.array(y_test)


  # Begin building the model
  from keras.models import Sequential
  from keras.layers import LSTM
  from keras.layers import Dense
  from keras.layers import Flatten, Dropout
  from keras.metrics import Accuracy

  model=Sequential()
  # Add layer 1
  model.add(LSTM(units=250,
                return_sequences=True,
                input_shape=(X_train.shape[1], 1))) 
  model.add(Dropout(0.2))

  # Add layer 2
  model.add(LSTM(units=50))
  model.add(Dropout(0.2))

  # Add dense layer
  model.add(Dense(1))
  model.compile(optimizer='adam',
                loss='mse')
  model.summary()

  # Fit model
  history = model.fit(X_train, 
                      y_train, 
                      epochs=20,
                      batch_size=32)
  # Print loss table
  print('___________________')
  print(' Epoch ', ' Loss ')
  print('-------------------')
  e=1  # starting epoch
  for i in range(len(history.history['loss'])):
    print('%s       '%e + str(round(history.history['loss'][i],4)))
    e=e+1   # update epoch


  # Make predictions
  y_test_pred=model.predict(X_test)

  # Predictions need to be transformed back to original scale
  y_test_pred_actual=scaler.inverse_transform(y_test_pred)

  # Plot the model and predictions
  final_df=df.iloc[:X_test.shape[0]]
  final_df['pred'] = y_test_pred_actual

  fig, [ax1, ax2] = plt.subplots(ncols=1, nrows=2, figsize=(10,10))
  # Metrics plot
  loss_per_epoch = history.history['loss']
  ax1.plot(range(len(loss_per_epoch)), loss_per_epoch)

  ax1.set_ylabel('Loss')
  ax1.set_xlabel('Epoch')
  ax1.set_title('Model Loss (hops = %s MSE: %s)'%(hops,round(loss_per_epoch[-1],6)))

  ax2.scatter(final_df['Decimal-Year'],
              final_df['UD(cm)'],
              color='red',
              label='actual',
              s=3)
  ax2.scatter(final_df['Decimal-Year'],
              final_df['pred'],
              color='blue',
              label='predicted',
              s=3)
  ax2.legend(loc='best')
  ax2.set_title('Actual vs. Predictions')
  ax2.set_xlabel('Decimal Years')
  ax2.set_ylabel('Z (cm)')
  print('Final Mean Square Error: %s'%loss_per_epoch[-1])

  return

# Loop over 1 week to 4 weeks for predictions
hop_list = [7, 14, 21, 28]
for n in hop_list:
  print('Days: %s'%n)
  univariate_lstm(df, n_days=n)
```

# Multivariate LSTM model (Bringing in time and position)

```python
# Visualize the dataset
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df['Decimal-Year'],
        df['UD(cm)'],
        'k.')
ax.set_xlabel('Time (day)')
ax.set_ylabel('Z (cm)')
print(df.head())

```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Flatten, Dropout
from keras.metrics import Accuracy


def multivariate_lstm(df, n_days):
  test_size=int(df.shape[0] * 0.3)
  df_train=df.iloc[0:test_size,:]
  df_test=df.iloc[test_size:,:]
  print(df_train.head())
  # Scale the data and this time just use the regular dataframe
  sc=MinMaxScaler()
  df_train_scaled=sc.fit_transform(df_train)

  # We need 2 different scaler objects. 
  # One for the input and one for the output
  sc2=MinMaxScaler()
  df_train_scaled_y=sc2.fit_transform(df_train[['UD(cm)']])

  hops=n_days
  no_records=df_train_scaled_y.shape[0]
  no_cols=2
  X_train=[]
  y_train=[]

  for i in range(hops,no_records):
    X_train.append(df_train_scaled[i-hops:i])
    y_train.append(df_train_scaled_y[i][0])

  X_train, y_train=np.array(X_train), np.array(y_train)

  # Build model
  model=Sequential()
  model.add(LSTM(units=250,
                return_sequences=True,
                input_shape=(hops,2)))
  model.add(Dropout(0.2))
  model.add(LSTM(units=50))
  model.add(Dropout(0.2))
  model.add(Dense(1))
  model.compile(optimizer='adam',
                loss='mean_squared_error')
  model.summary()

  # Fit the model
  history = model.fit(X_train,
                      y_train,
                      epochs=20,
                      batch_size=32)
  
    # Print loss table
  print('___________________')
  print(' Epoch ', ' Loss ')
  print('-------------------')
  e=1  # starting epoch
  for i in range(len(history.history['loss'])):
    print('%s       '%e + str(round(history.history['loss'][i],4)))
    e=e+1   # update epoch

  # Prepare test dataset  
  df_train_last14=df_train.iloc[-hops:]  # Take last batch
  df_test_full=df_test
  full_df=pd.concat((df_train_last14, df_test_full),axis=0)

  # Rescale the test dataframe
  full_df=sc.transform(full_df)

  # Reshape the test dataframe the same way as train
  no_record=full_df.shape[0]
  no_cols=2
  X_train_shape_pred=[]

  for i in range(hops, no_record):
    X_train_shape_pred.append(full_df[i-hops:i])
  X_train_shape_pred=np.array(X_train_shape_pred)

  # Scale back to original value to make predictions
  ytest=model.predict(X_train_shape_pred)
  y_final_pred=sc2.inverse_transform(ytest)

  fully_final=df_test
  fully_final['pred'] = y_final_pred

  fig, [ax1, ax2] = plt.subplots(ncols=1, nrows=2, figsize=(10,10))
  # Metrics plot
  loss_per_epoch = history.history['loss']
  ax1.plot(range(len(loss_per_epoch)), loss_per_epoch)

  ax1.set_ylabel('Loss')
  ax1.set_xlabel('Epoch')
  ax1.set_title('Model Loss (hops = %s, MSE: %s)'%(hops,round(loss_per_epoch[-1],6)))

  # Predictions
  ax2.scatter(fully_final['Decimal-Year'],
              fully_final['UD(cm)'],            
              label='actual',
              color='red',
              s=3)
  ax2.scatter(fully_final['Decimal-Year'],
              fully_final['pred'],
              label='pred',
              color='blue',
              s=3)
  ax2.legend(loc='best')
  ax2.set_title('Actual vs. Predictions')
  ax2.set_xlabel('Decimal Years')
  ax2.set_ylabel('Z (cm)')
  print('Final Mean Square Error: %s'%round(loss_per_epoch[-1],6))

  return


# Loop over 1 week to 4 weeks for predictions
hop_list = [7, 14, 21, 28]
for n in hop_list:
  print('Days: %s'%n)
  multivariate_lstm(df, n_days=n)
```
